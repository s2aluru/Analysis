## Basic Satellite Image Analysis

### In this notebook, we will analyze satellite image of an agricultural land and find its altitude and slope

```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
%matplotlib inline
from skimage import data
import warnings
warnings.filterwarnings('ignore')
```

```python
imagePath = "Datasets\\agriculture.jpg"
photo_data = misc.imread(imagePath)
plt.imshow(photo_data);
```

```python
type(photo_data)
photo_data.shape
```

The shape of the ndarray show that it is a three layered matrix. The first two numbers here are length and width, and the third number (i.e. 3) is for three layers: Red, Green and Blue.
<p style="font-family: Arial; font-size:1.25em;color:#2462C0; font-style:bold">
RGB color mapping</p> 
<ul>
<li><p style="font-family: Arial; font-size:1.0em;color:red; font-style:bold">
RED pixel indicates Altitude</p>
<li><p style="font-family: Arial; font-size:1.0em;color:green; font-style:bold">
GREEN pixel indicates Slope
<li><p style="font-family: Arial; font-size:1.0em;color:blue; font-style:bold">
BLUE pixel indicates Aspect (An aspect is the compass direction that a slope faces)
</ul>
</p>
The higher values denote higher altitude, aspect and slope.

```python
# Let's define color masks
red_mask   = photo_data[:, : ,0] < 150
green_mask = photo_data[:, : ,1] < 150
blue_mask  = photo_data[:, : ,2] < 150
```

Let us apply red mask and highlight all the high altitude areas. We will do this by detecting high intensity RED Pixels and muting down other areas.

```python
photo_red = misc.imread(imagePath);
photo_red[red_mask] = 0
plt.imshow(photo_red);
```

Looks like this land is on high altitude. Let us apply green mask and highlight all the high slope areas.

```python
photo_green = misc.imread(imagePath);
photo_green[green_mask] = 0
plt.imshow(photo_green);
```

Looks like part of the land is slopy. Let us apply blue mask and highlight all the high aspect areas. For a basic mask, it will be pretty much same as slope.

```python
photo_blue = misc.imread(imagePath);
photo_blue[green_mask] = 0
plt.imshow(photo_blue);
```
